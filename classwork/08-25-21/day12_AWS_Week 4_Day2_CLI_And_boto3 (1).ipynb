### Today's Lesson (Hands On)

Let's get your envionment set up for Python development on AWS!

**Note:** the below steps are purposefully left vague so you can get hands-on expereince and try it out for yourself.

1. Create IAM Admin user
2. Set up MFA for root user (and also admin user)
3. Create access keys for admin user, save them somewhere safe.
4. Download VS Code or PyCharm to use as an IDE.
5. Download the AWS CLI.
6. Configure the AWS CLI.
7. Common AWS CLI commands using S3: mb, cp, mv
8. Install boto3.
9. Go to Quickstart guide for boto3 and run a basic scripy to verify it works (see S3 example)
10. Create an S3 bucket using Boto3.
11. Create a Python function using Boto3 and call the function.
12. Create an EC2 instance using Boto3 and shut it down.
13. Create a cronjob to run your Boto3 Python script everyday at 7am.
14. [Boto3 challenge](https://github.com/BigFootAlchemy/APIChallenge) Try #1 & #3
15. [Boto3 cheatsheet](https://towardsdatascience.com/working-with-amazon-s3-buckets-with-boto3-785252ea22e0)
16. [More Boto3 practice](https://github.com/musayev-io/boto3-practice)
